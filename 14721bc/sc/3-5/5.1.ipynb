```python
A = [[5,0.02],[10,0.2],[20,0.4],[25,0.7],[30,0.9],[40,1]]
```

```python
B = [[10,0.4],[20,0.1],[25,0.9],[30,0.2],[40,0.6]]
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(A, columns = ['set/U', 'MembershipA']) 
df1
```

```python
df2 = pd.DataFrame(B, columns = ['set/U', 'MembershipB']) 
df2
```

```python
def kunion(df1,df2):
    dft = pd.merge(df1, df2, on = "set/U", how = "outer")
    union=[] #and operation
    for i in range(len(dft)):
        if (dft.at[i,'MembershipA'])<(dft.at[i,'MembershipB']):
            union.append(dft.at[i,'MembershipB'])
        else:
            union.append(dft.at[i,'MembershipA'])
    dft['union'] = union
    return dft
```

```python
kunion(df1,df2)
```

```python
def kintersection(df1,df2):
    dft = pd.merge(df1, df2, on = "set/U", how = "inner")
    intersection=[]
    for i in range(len(dft)):
        if (dft.at[i,'MembershipA'])>(dft.at[i,'MembershipB']):
            intersection.append(dft.at[i,'MembershipB'])
        else:
            intersection.append(dft.at[i,'MembershipA'])
    dft['intersection'] = intersection
    return dft
```

```python
kintersection(df1,df2)
```

```python
def difference(df1,df2):
    dft2 = pd.merge(df1, df2, on = "set/U", how = "outer")
    difference=[]
    complement2=[]
    for i in range(len(df1)):
        complement2.append(1-(dft2.at[i,'MembershipB']))
        if (dft2.at[i,'MembershipA'])>(complement2[i]):
            difference.append(complement2[i])
        else:
            difference.append(dft2.at[i,'MembershipA'])
    dft2['difference'] = difference
    return dft2
```
