Flow shop[non premptive]

```python
process = [[2, 3, 5],[0, 3, 2]] 

po=[]#processor occupied
ti1=[]#task in progress for process 1
ti2=[]#task in progress for process 2
tt=[]#total time

for i in range (len(process[0])):
    po.append(9)
#po[0]=1
print(po)

ti=[[0, 0, 2],[0, 0, 2]] #write code like po for this
print(ti)

for list in process:
    for number in list:
        tt.append(number)
#print(tt)
total=0
for ele in range(0, len(tt)): 
    total = total + tt[ele] 

#printing total value 
print("Total: ", total) 
#non premptive = 11
#premptive = 12
```

```python
def flowshop():
    time = 0
    x=0
    y=0
    
    for time in range(total):
        print("\n")
        print("\n")
        print(f"time=",time)
        count=0
        for x in range(len(process)):
            print("\n")
            print(f"x={x}")
            for y in range(len(process[0])):
                print(f"y={y}")
                print("po=",po)
                print("ti=",ti)
                print(f"process={process}")
                print(f"dealing with process: process[{x}],[{y}]")
                if process[x][y]!=0:
                    if po[y]==9 and ti[x][y-1]==2:
                        print(f"process[{x}][{y}] before={process[x][y]}")
                        process[x][y] = process[x][y] - 1
                        print(f"process[{x}][{y}] after={process[x][y]}")
                        ti[x][y]=1
                        print("ti=",ti)
                        print(process)
                        po[y]=x
                        print(f"po[{y}]={po[y]}")
                        if process[x][y]==0:
                            print(f"process task [{x}],[{y}] has just completed")
                            po[y]=9
                            ti[x][y]=2
                            break
                    
                    elif po[y]==x and ti[x][y-1]==2:
                        print(f"process[{x}][{y}] before={process[x][y]}")
                        process[x][y] = process[x][y] - 1
                        print(f"process[{x}][{y}] after={process[x][y]}")
                        print(process)
                        #po[y]=x
                        print(f"po[{y}]={po[y]}")
                        if process[x][y]==0:
                            print(f"process task [{x}],[{y}] has just completed")
                            po[y]=9
                            ti[x][y]=2
                            break
                    else:
                        print("nothing executed")
                else:
                    print("task already completed")
                    ti[x][y]=2
                    count = count+1
        if count==len(tt):
            print("time required",time)
            break
```
