```python
def flowshop():
    time = 0
    x=0
    y=0
    
    for time in range(total):
        print("\n")
        print("\n")
        print(f"time=",time+1)
        count=0
        for x in range(len(process)):
            print("\n")
            print(f"x={x}")
            for y in range(len(process[0])):
                print(f"y={y+1}")
                print("po=",po)
                print("ti=",ti)
                print(f"process={process}")
                print(f"dealing with process: process[{x+1}],[{y+1}]")
                if process[x][y]!=0:
                    if po[y]==9 and ti[x][y-1]==2:
                        print(f"process[{x+1}][{y+1}] before={process[x][y]}")
                        process[x][y] = process[x][y] - 1
                        print(f"process[{x+1}][{y+1}] after={process[x][y]}")
                        ti[x][y]=1
                        print("ti=",ti)
                        print(process)
                        gprocess.append(x)
                        gtask.append(y)
                        gtime.append(time)
                        po[y]=x
                        print(f"po[{y+1}]={po[y]+1}")
                        if process[x][y]==0:
                            print(f"process task [{x+1}],[{y+1}] has just completed")
                            po[y]=9
                            ti[x][y]=2
                            break
                    
                    elif po[y]==x and ti[x][y-1]==2:
                        print(f"process[{x+1}][{y+1}] before={process[x][y]}")
                        process[x][y] = process[x][y] - 1
                        print(f"process[{x+1}][{y+1}] after={process[x][y]}")
                        print(process)
                        gprocess.append(x)
                        gtask.append(y)
                        gtime.append(time)
                        #po[y]=x
                        print(f"po[{y+1}]={po[y]}")
                        if process[x][y]==0:
                            print(f"process task [{x+1}],[{y+1}] has just completed")
                            po[y]=9
                            ti[x][y]=2
                            break
                    else:
                        print("nothing executed")
                else:
                    print(f"task {x+1},{y+1} is already completed")
                    ti[x][y]=2
                    count = count+1
        if count==len(tt):
            print("time required",time)
            break
```
