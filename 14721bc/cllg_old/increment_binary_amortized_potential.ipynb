```python
nstr=input("enter the counte value") 
n=int(nstr)
```

```python
def decimalToBinary(n):  
    return bin(n).replace("0b", "")  
```

```python
def countX(lst, x): 
    count = 0
    for ele in lst: 
        if (ele == x): 
            count = count + 1
    return count 
```

```python
bn=[0]
ac=[0]
pd=[0]
fidi=[0]
amtc=[0]
tamtc=0
tac=0
```

```python

for i in range(1,n+1):
    acc=0
    fidic=0
    b1=decimalToBinary(i)
    b0=decimalToBinary(i-1)
    bl1=list(b1)
    #print("bl1=",bl1)
    bl0=list(b0)
    #print("bl0=",bl0)
    bl0 = list(map(int, bl0))
    bl1 = list(map(int, bl1))
    
    if len(bl1)!=len(bl0):
        #d=len(bl1)-len(bl0)
        bl0.insert(0,0)
    
    print("bl1=",bl1)
    print("bl0=",bl0)
    fidic=countX(bl1,1)
    print("fidic=",fidic)
    for j in range(0,len(bl1)):
        print("j=",j)
        print("len(bl1)=",len(bl1))
        print("bl1[j]")
        if bl1[j]!=bl0[j]:
            acc=acc+1
            print("acc=",acc)
            continue

    print("---")
    ac.append(acc)
    print("ac=",ac)
    fidi.append(fidic)
    print("fidi=",fidi)
    pdr=fidi[i]-fidi[i-1]
    print("pdr=",pdr)
    pd.append(pdr)
    print("pd=",pd)
    acr=ac[i]+pd[i]
    print("acr=",acr)
    amtc.append(acr)
    print("amtc=",amtc)
    print("***********")
for ele in range(0, len(amtc)): 
    tamtc = tamtc + amtc[ele]
for ele in range(0, len(ac)): 
    tac = tac + ac[ele]
print("tamtc=",tamtc)
print("tac=",tac)
print("\n\n")
```
