```python
from sympy import Symbol, Derivative
from scipy import optimize
import numpy as np
from sympy import *
```

```python
def f(v,t):
    return abs((((8.3144621*10**-5)*t)/(v-0.08664))-(0.4275/((t**0.5)*v)*(v+0.08664)))
#     R=8.3144621(75) × 10−5 m3 bar K−1 mol−1
```

```python
t1=f(2,373.51)
```

```python
v= Symbol('v')
t=373.51
function= (((8.3144621*10**-5)*t/(v-0.08664))-(0.4275/((t**0.5)*v)*(v+0.08664)))
deriv= Derivative(function, v)
```

```python
t2=abs(deriv.doit().subs({v:2}))
```

```python
def nrm(x1,t):
    
#     #code for derivation
#     v= Symbol('v')
#     function= (((8.3144621*10**-5)*t/(v-0.08664))-(0.4275/((t**0.5)*v)*(v+0.08664)))
#     deriv= Derivative(function, v)

    #newton raphson method
    t1=f(2,373.51)
    t2=abs(deriv.doit().subs({v:2}))
    x2 = x1-(t1/t2)
    print(x2)
    if abs(x2-x1)>0.0001:
        #print(x1)
        x1=x2
        nrm(x1,t)
    else:
        print("root of eqn is ",x2)
        return x2
```
