```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
#make test list non xl
nxl=[]
for c in range(1,len(xl2)-1,2):
    nxl.append(xl2[c])
nxl
```

```python
def ylm(x): # y / f(x)
    return 1/((x**2)+1)
def dsm(xl): #y list / data set maker
    yl=[]
    for i in range(len(xl)):
        yl.append(ylm(xl[i]))
    return yl
```

```python
yl=dsm(xl)
yl
```

```python
def lmm(yl): #newton matrix maker
    lm=[yl]
#     lm.insert(0, number_tuple)
    for i in range(1,len(yl)):
        tl=[]
        for j in range(len(lm[i-1])-1):
#             print("i=",i)
#             print("j=",j)
            t=(lm[i-1][j+1]-lm[i-1][j])
#             print(t)
            tl.append(t)
        lm.append(tl)
    for k in range(1,len(xl)):
        t=np.zeros(k)
        lm[k].extend(t)
    lm.append(np.zeros(len(xl)))
    return lm
```

```python
lm=lmm(yl)
```

```python
def pm(lm,xl,yl,x): #polynomial maker
    h=xl[2]-xl[1]
#     print(h)
    u=(x-xl[0])/h  #newton forward
#     print(u)
    ys=len(xl)-1 #y0 subscript
    for _ in range(1,len(xl)):
        if x<xl[(len(xl)-_)]:
#             print(f"{x}<{xl[(len(xl)-_)]}")
            ys=ys-1
#             print(ys)
#     ys=ys
#     print("\n")
#     print(ys)
    p=yl[ys]
#     print("p=",p)
    
    pl=[(yl[ys-1])]
    
    for l in range((len(xl)-ys)):
#         print("\n")
#         print("l=",l+1)
        yt=lm[l+1][ys]
        ut=u
        for ll in range(0,l+1):
            if ll>0:
                ut=ut*(u-ll)
#         print("ut=",ut)
#         print("yt=",yt)
#         print("l=",l+1)
#         print("fact=",math.factorial(l+1))
        
        p=p+((ut*yt)/math.factorial(l+1))
        pl.append(p)
#         print("p=",p)
#         print("\n")
    return p
```

```python
pm(lm,xl,yl,2.75)
```

```python
apl=[] # actual polynomial list
fxl=[] # f(x) list
for cc in range(len(nxl)):
#     print(pm(lm2,xl,yl,cc))
    tv=nxl[cc]
#     print(tv)
    apl.append(pm(lm,xl,yl,tv))
    fxl.append(ylm(tv))
```

```python
#calculating discrepancy
disc=[]
for fl in range(len(fxl)):
    disc.append(abs(fxl[fl]-apl[fl]))
```
