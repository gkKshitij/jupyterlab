```python
import math
#Bisection method
r=1
d=2
#for 20%
R=1
print("\n\nPart 1: For 20%")
def printIt(y,a,b):
    print("\nGiven two equations: \n1. R^2.cos^-1(R-h/R)-(R-h)(R*h - h^2)^0.5\n2. A - 0.2*pi*R*R = 0\n R=1\n D=2")
    print("\nLet's assume two values as our initial guesses for h as {} and {}".format(a,b))
    print("\nFor h={}".format(a))
    R=1.0
    h=a
    A=((R**2)*math.acos(((R-h)/R)))-((R-h)*(((R*h)-(h**2))**0.5))
    eqn=(A - (y*3.14*R*R))
    print("\n A = %.4f" % A)
    print(" eqn = %.4f" % eqn)
    print("For h={}".format(b))
    R=1.0
    h=b
    A=((R**2)*math.acos(((R-h)/R)))-((R-h)*(((R*h)-(h**2))**0.5))
    eqn=(A - (y*3.14*R*R))
    print("\n A = %.4f" % A)
    print(" eqn = %.4f" % eqn)
    print("\nTherefore, \n a={}\n b={}".format(a,b))

def solveIt(y,a,b):

    i=1
    while True:
        print("\nIteration {}".format(i))
        i=i+1
        x=(a+b)/2
        h=x
        A=((R**2)*math.acos(((R-h)/R)))-((R-h)*(((R*h)-(h**2))**0.5))
        eqn=(A - (y*3.14*R*R))
        print(" a = %.4f" % a)
        print(" b = %.4f" % b)
        print(" x = %.4f" % x)
        print(" A = %.4f" % A)
        print(" eqn = %.4f" % eqn)
        if (round(eqn, 4)<0.0):
            a=x
        elif (round(eqn, 4)>0.0):
            b=x
        elif (round(eqn, 4)==0.0):
            print("\nThe answer is: %.4f" %x)
            print("\nh\d is: %.4f" %(x/2))
            break

printIt(0.2,0.35,0.4)
solveIt(0.2,0.35,0.4)

#for 40%
print("\nPart 2: For 40%")
printIt(0.4,0.75,0.8)
solveIt(0.4,0.75,0.8)

#for 60%
#print("\nPart 3: For 60%")
#printIt(0.6,0,1)
#solveIt(0.6,0,1)

#for 80%
#print("\nPart 4: For 80%")
#printIt(0.8,0,1)
#solveIt(0.8,0,1)

#print(" eqn = %.4f" % eqn)
#print(round(12.54673, 2))
```
