```python
from scipy import misc
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = 1.1
```

```python
def f(x):
    return (1.0/1.0)*x**2-(2.0/1.0)*x**1+(1.0/1.0)
```

```python
step = []
xk = []
ek = []
fxk = []
cr = []
```

```python
def NewtonsMethod(f, x, tolerance=10**-12):
    z=0
    while True:
        #print(x)
        
        #print("iteration number",z)
        x1 = x - (f(x) / misc.derivative(f, x))
        t = abs(x1 - x)
        
        #convergence ratio
        crc=(x1-1)/(x-1)

        #the tabulation code
        step.append(z+1)
        xk.append(x)
        ek.append(t)
        fxk.append(f(x))
        cr.append(crc)
        
        if t < tolerance:
            x=x1
#             step.append(z+1)
#             xk.append(x)
#             ek.append(t)
#             fxk.append(f(x))
            break
        
        x = x1
        z=z+1
        
    return x,z;
```

```python
x0,z = NewtonsMethod(f, x)
```

```python
print('x: ', x)
print('x0: ', x0)
print("f(x0) = ", (1.0/1.0)*x**2-(2.0/1.0)*x**1+(1.0/1.0))
print("number of iteration=",z+1)
```

```python
df = pd.DataFrame(list(zip(step,xk,fxk,ek,cr)), 
               columns =['step', 'x(k)', 'f(x(k))', 'e(k)','cr']) 
```

```python
def f2(x):
    return (1.0/1.0)*x**2-(3.0/1.0)*x**1+(2.0/1.0)
```

```python
step2 = []
xk2 = []
ek2 = []
fxk2 = []
cr2 = []
```

```python
def NewtonsMethod2(f, x, tolerance=10**-12):
    z=0
    while True:
        #print(x)
        
        #print("iteration number",z)
        x1 = x - f(x) / misc.derivative(f, x) 
        t = abs(x1 - x)
        
        #convergence ratio
        crc=(x1-1)/(x-1)
        
        #the tabulation code
        step2.append(z+1)
        xk2.append(x)
        ek2.append(t)
        fxk2.append(f(x))
        cr2.append(crc)
        
        if t < tolerance:
            break
        x = x1
        z=z+1
        
        
    return x,z;
```

```python
x0,z = NewtonsMethod2(f2, x)
```

```python
print('x: ', x)
print('x0: ', x0)
print("f(x0) = ", (1.0/1.0)*x**2-(3.0/1.0)*x**1+(2.0/1.0))
print("number of iteration=",z+1)
```

```python
df2 = pd.DataFrame(list(zip(step2,xk2,fxk2,ek2,cr2)), 
               columns =['step', 'x(k)', 'f(x(k))', 'e(k)','cr']) 
```
